```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(file_path):
    """Load the unemployment dataset."""
    df = pd.read_csv(file_path)
    return df

def preprocess_data(df):
    """Clean and preprocess the data."""
    df.dropna(inplace=True)  # Remove missing values if any
    df.columns = df.columns.str.strip()  # Remove leading/trailing spaces from column names
    df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)  # Standardize column names
    return df

def explore_data(df):
    """Display basic dataset information."""
    print("\nDataset Information:")
    print(df.info())
    print("\nSummary Statistics:")
    print(df.describe())
    print("\nMissing Values:")
    print(df.isnull().sum())

def visualize_data(df):
    """Generate visualizations for unemployment trends."""
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=df['date'], y=df['estimated_unemployment_rate_(%)'], marker='o')
    plt.xticks(rotation=45)
    plt.title("Unemployment Rate Over Time in India")
    plt.xlabel("Date")
    plt.ylabel("Unemployment Rate (%)")
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df['estimated_unemployment_rate_(%)'], bins=20, kde=True)
    plt.title("Distribution of Unemployment Rate in India")
    plt.xlabel("Unemployment Rate (%)")
    plt.show()

def main():
    file_path = "C:/Users/Manasvi/OneDrive/Desktop/Unemployment in India.csv"  # Updated with actual file path
    
    # Load and process data
    df = load_data(file_path)
    df = preprocess_data(df)
    
    # Explore dataset
    explore_data(df)
    
    # Visualize unemployment trends
    visualize_data(df)
    
if __name__ == "__main__":
    main()
```
